### 1.2. Изучите стоимости недвижимости

```python
import numpy as np 
import matplotlib.pyplot as plt 
plt.figure(figsize=(6,4))
plt.hist(df['price'],bins='auto',log=True)
plt.title("Распределение стоимости")
plt.xlabel("Стоимость, млн.")
plt.ylabel("Количество")
plt.xticks(rotation=90)
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
plt.figure(figsize=(6,4))
plt.hist(df ['sqft_living'],bins='auto',log=True)
plt.title("Распределение жилой площади")
plt.xlabel("Кв.м.")
plt.ylabel("Количество, ед.")
plt.xticks(rotation=90)
```

### 1.4. Изучите распределение года постройки

```python
plt.hist(df['yr_built'], bins=range(df['yr_built'].min(), df['yr_built'].max() + 1))
plt.xlabel("Год постройки")
plt.ylabel("Количество домов, построенных за год, ед.")
plt.xticks(rotation=0)
```

### Зависимость цены жилья от жилой площади

```python
import seaborn as sns
sns.jointplot(x = df ['price'], y = df ['sqft_living'],kind = 'reg')
plt.title("Зависимость цены жилья от жилой площади")
plt.xlabel("Цена дома, млн")
plt.ylabel("Жилая площадь, кв.м.")
plt.xticks(rotation=0)
```

Вывод: на графике можно заметить сильную прямую зависимость цены жилья от его площади, особенно в нижнем ценовом диапазоне. Для домов с большей площадью наблюдается более существенный разброс цен.

### Зависимость цены от этажности дома

```python
data1 = df.groupby('floors').agg({'price': 'mean'}).reset_index()
plt.bar(data1['floors'], data1['price'])
plt.title("Зависимость цены дома от его этажности")
plt.xlabel("Количество этажей")
plt.ylabel("Средняя цена дома, млн");
plt.xticks(rotation=0)
```

Вывод: Для домов с этажностью с 1 по 2.5 этажа наблюдается равномерный рост цены. 

### Зависимость цены дома от вида

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('Цена дома, млн')
plt.ylabel('Вид')
plt.title('Зависимость цены дома от вида');
```

Вывод: наибольшую среднюю стоимость имеют дома с оценкой вида 4.

### Зависимость цены дома от количества спален

```python
data2 = df.groupby('bedrooms').agg({'price': 'mean'}).reset_index()
plt.bar(data2['bedrooms'], data2['price'])
plt.title("Зависимость цены дома от количества спален")
plt.xlabel("Количество спален")
plt.ylabel("Средняя цена дома, млн");
plt.xticks(rotation=0)
```

Вывод: самые высокие средние цены на дома с количеством спален от 5 до 10. 

### Зависимость цены жилья от оценки состояния дома

```python
sns.boxplot(x = df ['price'], y = df ['condition'].astype('str'), whis = 1.5)
plt.title("Зависимость цены жилья от оценки состояния дома")
plt.xlabel("Цена дома, млн")
plt.ylabel("Оценка состояния дома")
plt.xticks(rotation=0)
```

Вывод: на графике можно увидеть, что средняя стоимость домов с оценкой от 3 до 5 выше, чем с оценкой состояния от 1 до 2.
